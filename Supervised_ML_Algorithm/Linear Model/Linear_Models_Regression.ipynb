```python
# Here Introduce 3 linear models for Regression
# 1. Linear Regression
# 2. Ridge Regression
# 3. Lasso
#
# First, introduce the regression.  Regression use the following form to make the prediction.
#       y = w[0] * x[0] + w[1] * x[1] + w[2] * x[2] + .... + w[p] * x[p] + b
#                     x[n] n=0 ~ p, denote the features of a single data point
#                     w, b are the parameters that model has learned from training data
#                     y is the predictoin that the model make
# When there is only one feature, the equation become a line equation.
#       y = w[0] * x[0] + b
#
# We can use the one-diamension dataset, wave, to learn the parameters, w and b.

import mglearn
```

```python
mglearn.plots.plot_linear_regression_wave()
```

```python
# Comparing the prediction made by straight line with KNeighborsRegressor, 
# using a straight line the make prediction seems very restrictive.
# It is a strong and somehow unrealistic assumption that the target y is a linear combination of features.
# But, remember that the parameters, w and b, are learned according to one feature.
# It is just a one diamension data.  And it is not bery fair to see by one diamension data.
#  ***** In fact, for the dataset with many features, the linear models can be very powerful. *****
#
#
# There are many different linear models for the regression.  The differences are
#        1. How the model learned the parameters, w and b.
#        2. How the model complexity can be controlled.
#
#
# Linear Regression (aka ordinary least squares): the most popula linear model for regression
# It finds w and b that minimize the mean square error between the prediction and the true target yon the training dataset.
#                   Mean Squared Error: the sum of squared differences between the prediction and the true value.
# Linear Regression has no parameter, which is a benifut, however is has no way to control the complexity.

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LinearRegression().fit(X_train, y_train)
# coefficient and intercept are learned from the training dataset
print('Coefficient: {}'.format(lr.coef_))
print('Intercept: {}'.format(lr.intercept_))
```

```python
# Find the performance of training and test data
print('The score of training data: {:.2f}'.format(lr.score(X_train, y_train)))
print('The socre of test data: {:.2f}'.format(lr.score(X_test, y_test)))
```

```python
# The scores of train and test dataset are very close.
# It means that they are underfitting, not overfitting.
# Since the wave is a one dimension dataset, it is very simple and has small chance to overfitting.
# However, with higher-dimension datasets, linear model become very powerful.
# And it has a higher of chances of overfitting.
#
#
# Try the Boston House dataset, it has 506 samples and 105 derived features

X, y = mglearn.datasets.load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
# Create the linear regression object and train it by the training data
lr = LinearRegression().fit(X_train, y_train)

# Show the parameters, w and b
print('Coefficient: {}'.format(lr.coef_))
print('Interception: {:.2f}'.format(lr.intercept_))

# Show the R^2 of training and test data
print('Performance of training data: {:.2f}'.format(lr.score(X_train, y_train)))
print('Performance of test data: {:.2f}'.format(lr.score(X_test, y_test)))
```

```python
# The difference of performances indicates that this moder is overfitting.
# We want to find a model to control the complexity
# We try 'Ridge Regression'
```

```python
# Ridge Regression is one of the linear regression model.
# All what ridge to do is to avoid overfitting.
# So, the key point of ridge regression is that we try to minimize the value of the coeifficients.  -> All w should close to 0.
# All features should have as less effect on the outcome as possible, while the mode still preidcting well.
# This process is so called regularization.  Ridge uses L2 Regularization (a mean square approach).

from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train, y_train)
print('Ridge Regression, Training dataset score: {:.4f}'.format(ridge.score(X_train, y_train)))
print('Ridge Regression, Test dataset score: {:.4f}'.format(ridge.score(X_test, y_test)))
```

```python
# Comparing the scores of LInearRegression and Ridge models:
# LinearRegression, Training dataset score: 0.95
# LinearRegression, Test datset socre: 0.61
# Ridge, Training dataset socre: 0.89
# Ridge, Test dataset score: 0.75
# Training dataset score: LinearRegression is higher than  Ridge
# Test dataset score:      Ridge                 is higher than  LinearRegression
#
# Summary: Ridge is less overfitting and is better in predicting test dataset.
#                Ridge is less complex than LinearRegression. -> Ridge performance is poor in training but good in test dataset.
#                Weighting: Ridge weighting the complexity and training dataset performance by the parameter, alpha.
#                                Adjust alpha to get the optimum setting in different dataset.
#                                Incresing alpha forces coeifficients to move to 0, which cause less complexity and poor performance for training data,
#                                but increasing the generalization(good for test data performance).
#
# Test different value of alpha.

ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print('Ridge, alpha=10, training dataset score: {:.4f}'.format(ridge10.score(X_train, y_train)))
print('Ridge, alpha=10, test dataset score: {:.4f}'.format(ridge10.score(X_test, y_test)))

ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print('Ridge, alpha=0.1, training dataset score: {:.4f}'.format(ridge01.score(X_train, y_train)))
print('Ridge, alpha=0.1, test dataset score: {:.4f}'.format(ridge01.score(X_test, y_test)))
```

```python
# Alpha                     10          1          0.1        LinearRegression
# Training score       0.79     0.89       0.93       0.95
# Test score            0.64     0.75       0.77        0.61
```

```python
# Show the relationship between coefficients and alpha

import matplotlib.pyplot as plt

plt.plot(ridge.coef_, 's', label = 'Ridge alpha=1')
plt.plot(ridge10.coef_, '^', label = 'Ridge alpha=10')
plt.plot(ridge01.coef_, 'v', label = 'Ridge alpha=0.1')

plt.plot(lr.coef_, 'o', label = 'LinearRegression')
plt.xlabel('Coefficient index')
plt.ylabel('Coefficient magnitude')
plt.hlines(0, 0, len(lr.coef_))
plt.ylim(-25, 25)
plt.legend()
```
