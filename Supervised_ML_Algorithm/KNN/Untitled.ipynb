```python
# KNN Machine Learning Algorithm (K Nearest Neighbor)
# It may be  the simplest machine learn algorithm.
# To make a prediction of a new point(or a variable), the algorithm just finds the closest
# data point in the dataset -> the new point's nearest training point.
import mglearn

mglearn.plots.plot_knn_classification(n_neighbors=1)
```

```python
# Instead of considering the nearest neighbor (n_neighbor=1), we can consider any number of neighbor.  That is, we can set the K to any value.
# Followings show the K value to 3 and see how the algorithm works.
mglearn.plots.plot_knn_classification(n_neighbors=3)
```

```python
# The KNN model first finds 3 training points nearest to the new points and predict the new point according the majority of these 3 neighors.
```
